```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']#显示中文
```

```python
df1 = pd.read_csv("chap05-cellphone.csv")
```

```python
#df1_a = df1[df1["MSales"]> 0]
#df1_a=df1_a[df1_a["Comments"]> 0]
#df2 = df1[df1["MSales"]> 0][df1["Comments"] > 0]
df2 = df1[(df1["MSales"]> 0)&(df1["Comments"] > 0)] #取MSales和Comments大于0的数据
```

```python
df2["MSales"] = df2["MSales"].apply(np.log) #对MSales取对数
```

```python
df2["Comments"] = df2["Comments"].apply(np.log) #对Comments取对数
```

```python
# 1.基础统计图形
# 基础绘图（散点图）

plt.plot(df1['MSales'], df1['Comments'], color = "red", marker = "o", linestyle = "none") #指定颜色为红色
plt.show()
```

```python
# 添加图形元素的方式
#fig = plt.figure()
ax = plt.subplot(111)
ax.set(xlim=[0, 5000], ylim=[0, 5000], title="cellphone",
    ylabel="Comments", xlabel="MSales") #添加坐标轴名称和标题
plt.plot(df1['MSales'], df1['Comments'],"bo")
plt.show()
```

```python
# 参数和属性
plt.scatter('MSales', 'Comments', data = df2)
#添加坐标轴名称和标题
plt.title("Test")
plt.xlabel("MSales")
plt.ylabel("Comments")

plt.show()
```

```python
# 点参数
plt.scatter('MSales', 'Comments', data = df2[1:20],
    color = "red", #颜色
    marker = "+", #点型
    s = 200) #大小
plt.show()
```

```python
plt.scatter('MSales', 'Comments', data = df2[1:20],
    color = "red", #颜色
    marker = "o", #点型
    s = 100) #大小
plt.show()
```

```python
# 线参数
plt.plot('MSales', 'Comments', data = df2[1:20],
  linestyle = "-", #线型 
  linewidth = 1,  #宽度
  color = "red", #颜色
  marker = "o") #点型
plt.show()
```

```python
# 图例
plt.scatter('MSales', 'Comments', 
    data = df2[df2['Place'] == '广东 广州'][1:20],
    color = "red", marker = "o", label = "广州")
plt.scatter('MSales', 'Comments', 
    data = df2[df2['Place'] == '浙江 杭州'][1:20],
    color = "blue", marker = "o", label = "杭州")

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.legend(loc = "upper left")
plt.show()
```

```python
# 文字说明
plt.plot([1,2,3])
plt.text(1, 2, r'$\mu=100, \sigma=15$')
plt.show()
```

```python
plt.plot([1,2,3])
plt.annotate('test', xy=(1, 2), xytext=(1.5, 2.2),
    arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.show()
```

```python
# 2.可视化与数据分析
# 简单数据分析
df3 = df1[df1['Place'] == '广东 广州'].append(df1[df1['Place'] == '浙江 杭州'])
```

```python
print("mean=",np.mean(df3['MSales'])) #求MSales变量均值
print("std=",np.std(df3['MSales'])) #求MSales变量标准差
print("median=",np.median(df3['MSales'])) #求MSales变量中位数
print("percentile=",np.percentile(df3['MSales'], 25)) #求MSales变量分位数
```

```python
df3.groupby(by='Place').count() #根据Place分组，计算频数
```

```python
df3.groupby(by='Place').mean() #根据Place分组，计算均值
```

```python

plt.hist(df1['MSales'], bins = 50, density=True, histtype='bar')
plt.show()
```

```python
plt.hist(df2['MSales'], bins=20, density=True, histtype='bar')
plt.show()
```

```python
# 单变量的分布
# 画出连续型变量MSales的直方图
fig,ax = plt.subplots(1,2)
fig.set_size_inches(8,4)
ax[0].hist(df1['MSales'], bins = 50, density=True, histtype='bar')
ax[1].hist(df2['MSales'], bins=20, density=True, histtype='bar')
```

```python
# 画出离散型变量Place的条形图和饼图
dffreq = df3.groupby(by='Place').count()['Summary'] #频数分布表
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=False, sharey=False)
ax1.bar(np.arange(len(dffreq)), dffreq, color='lightblue', align='center') #条形图
ax2.pie(dffreq, labels = dffreq.index) #饼图
plt.show()
```

```python
# 两变量的关系

# 两个连续变量的关系：散点图

# 连续变量和离散变量的关系
# 画出变量MSales在不同Place组中的箱线图
v1 = df3[df3['Place'] == '广东 广州']['MSales']
v2 = df3[df3['Place'] == '浙江 杭州']['MSales']
plt.boxplot((v1, v2), labels=('广州','杭州'))
plt.show()
```

```python
# 两个离散变量的关系
# 画出马赛克图
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df3, ['Place', 'Protection'])
plt.show()
```
